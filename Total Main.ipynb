```python
df=pd.read_csv('eBM_data.csv')
Tag_list=[]
for tags in df.PROJECT_TAGS:
    taglist=tags.split(",")
    for tag in taglist:
        Tag_list.append(tag)
Tag_final_list=np.unique(Tag_list)
print(Tag_final_list)


df_svd = pd.DataFrame()
```

```python
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import random as sparse_random
project_id=df_svd["project_id"]
df_svd.drop(columns=["project_id"],inplace=True)
X=df_svd
svd = TruncatedSVD(n_components=50, n_iter=7, random_state=42)
svd.fit(X)
```
