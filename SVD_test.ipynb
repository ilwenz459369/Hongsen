```python
import pandas as pd
import numpy as np
import os
```

```python
df=pd.read_csv('eBM_cleaned_data.csv')
df.head()
```

```python
df.drop(columns=["SUPERVISOR","PROJECT_DESCRIPTION"],inplace=True)
df.head()
```

```python
# df_svd=df.groupby('supervisor_id').max()
df=df.sort_values(by = 'supervisor_id',axis = 0,ascending = True)
df.head()
```

```python
project_id=df["project_id"]
supervisor_id=df["supervisor_id"]
df.drop(columns=["project_id","supervisor_id"],inplace=True)
df_svd=df
df_svd.head()
```

```python
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import random as sparse_random

X=df_svd
svd = TruncatedSVD(n_components=50, n_iter=7, random_state=42)
svd.fit(X)
print(svd.explained_variance_ratio_.sum())
```

```python
print(svd.explained_variance_ratio_.sum())
```

```python
X_new=svd.fit_transform(X)
print(X_new)
```

```python
X_new=svd.fit_transform(X)
df_new=pd.DataFrame(X_new)
df_new["project_id"]=project_id
df_new["supervisor_id"]=supervisor_id
df_new.head()
```
